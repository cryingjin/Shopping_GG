# Data Load

```python
file_path_train = '../bigcontest_2020/2020빅콘테스트 문제데이터(데이터분석분야-챔피언리그)/01_제공데이터/'
file_path_test = '../bigcontest_2020/2020빅콘테스트 문제데이터(데이터분석분야-챔피언리그)/02_평가데이터/'

def file_load() : 
    print("\n Data Load....")
    
    sales = pd.read_excel(file_path_train+'2020 빅콘테스트 데이터분석분야-챔피언리그_2019년 실적데이터.xlsx'
                          , header=1 , parse_dates=['방송일시'])
    views = pd.read_excel(file_path_train+'2020 빅콘테스트 데이터분석분야-챔피언리그_시청률 데이터.xlsx'
                         , header=1, parse_dates=['시간대'])
    evals = pd.read_excel(file_path_test+'2020 빅콘테스트 데이터분석분야-챔피언리그_2020년 6월 판매실적예측데이터(평가데이터).xlsx'
                         , header=1, parse_dates=['방송일시'])
    
    print("\n Data Load complete !!")
    print("\n sales :", sales.shape)
    print("\n views :", views.shape)
    print("\n evals :", evals.shape)
    
    return(sales, views, evals)
```

```python
sales, views, evals = file_load()
```

### 시청률에 맞춰서 판매량도 변화하는가? 

```python
sales['주문량'] = round(sales['취급액']/sales['판매단가'])
```

```python
sales['year']= sales['방송일시'].map(lambda x: x.year)
sales['month'] = sales['방송일시'].map(lambda x: x.month)
sales['day'] = sales['방송일시'].map(lambda x: x.day)
sales['hour'] = sales['방송일시'].map(lambda x: x.hour)
sales['minute'] = sales['방송일시'].map(lambda x: x.minute)
sales['minute2'] = sales['minute'].map(lambda x: 1 if x>=30 else 0)

data = sales.groupby(['year','month','day','hour','minute2'])['주문량'].agg(lambda x: sum(x)/x.count()).reset_index()
data = data.groupby(['hour','minute2'])['주문량'].mean().reset_index()
data['hour'] = data['hour'].map(lambda x:"0"+str(x) if x<10 else str(x))
data['minute2'] = data['minute2'].map(lambda x: "00" if x == 0 else "30")

a = data[['hour','minute2']].apply(lambda x : str(x[0])+":"+ str(x[1]), axis=1)
data['times'] = a

five2 = pd.merge(five['times'], data[['times','주문량']], how='left', on='times')
five2 = five2.fillna(0)
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=five2
             , x='times'
             , y='주문량'
             )



plt.xticks(rotation=90)
plt.axvline(x='02:00', color='r', linestyle=':', linewidth=3)
plt.axvline(x='06:30', color='r', linestyle=':', linewidth=3)

plt.axvline(x='09:00', color='g', linestyle='-', linewidth=3) # 첫번째 지점 : 6시30분 ~ 9시 00분 (시청률 증감)
plt.axvline(x='12:30', color='y', linestyle='-', linewidth=3) # 두번째 지점 : 9시00분 ~ 12시30분 (시청률 폭발)
plt.axvline(x='20:00', color='b', linestyle='-', linewidth=3) # 세번째 지점 : 12시30분 ~ 20시00분 (시청률 폭발)
plt.show()
```

```python
data = sales[sales['주문량']<=1000].groupby(['year','month','day','hour','minute2'])['주문량'].agg(lambda x: sum(x)/x.count()).reset_index()
data = data.groupby(['hour','minute2'])['주문량'].mean().reset_index()
data['hour'] = data['hour'].map(lambda x:"0"+str(x) if x<10 else str(x))
data['minute2'] = data['minute2'].map(lambda x: "00" if x == 0 else "30")

a = data[['hour','minute2']].apply(lambda x : str(x[0])+":"+ str(x[1]), axis=1)
data['times'] = a

five2 = pd.merge(five['times'], data[['times','주문량']], how='left', on='times')
five2 = five2.fillna(0)
```

## 기본 변수 장착 

```python
# file reload
sales, views, evals = file_load() 

# NaN Preprocess
sales['노출(분)'] = sales['노출(분)'].fillna(method='pad') # nan 
sales['취급액'] = sales['취급액'].fillna(0)
#sales['노출(분)'] = sales['노출(분)'].map(lambda x: round(x,-1))

evals['노출(분)'] = evals['노출(분)'].fillna(method='pad') # nan
evals['취급액'] = evals['취급액'].fillna(0) # 취급액은 모르는 값 
#evals['노출(분)'] = evals['노출(분)'].map(lambda x: round(x,-1))

# Free product Preprocess 
sales = sales[sales['상품군']!='무형']

evals_free = evals[evals['상품군']=='무형']
evals = evals[evals['상품군']!='무형']

# error valeus Preprocess 
sales.loc[sales['취급액']==50000,'취급액'] = 0 

# Log y 
sales['log_y'] = np.log1p(sales['취급액'])
```

```python
def feature_engineering(train, test) : 
    
    train['label'] = 'train'
    test['label'] = 'test'

    ## times(노출 시간의 누적)
    train = train.sort_values(['상품명','방송일시'])
    test = test.sort_values(['상품명','방송일시'])    
    
    for i in range(1,train.shape[0]) : 
    
        if ( (train.iloc[i,0] - train.iloc[i-1,0] ).seconds <= 3600) & (train.iloc[i,4] == train.iloc[i-1,4]) : 
            train.iloc[i,1] = train.iloc[i,1] + train.iloc[i-1,1]

    for i in range(1,test.shape[0]) : 
    
        if ( (test.iloc[i,0] - test.iloc[i-1,0] ).seconds <= 3600) & (test.iloc[i,4] == test.iloc[i-1,4]) : 
            test.iloc[i,1] = test.iloc[i,1] + test.iloc[i-1,1]            
            
    
    ## born 
    y = train[['취급액','log_y']]
    data = pd.concat([train.drop(['취급액','log_y'], axis=1), test.drop(['취급액'],axis=1)], axis=0)
    
    ## feature 
    
    #1) times(노출 시간)
    data['month'] = data['방송일시'].map(lambda x: x.month)
    data['weekday'] = data['방송일시'].map(lambda x: x.weekday())
    data['hour'] = data['방송일시'].map(lambda x: x.hour) 
    
    data[['month','weekday','hour']] = data[['month','weekday','hour']].astype('category')
    
    
    # return 
    train = data[data['label']=='train']
    train = pd.concat([train, y], axis=1)
    test = data[data['label']=='test']
    
    return(train.drop(['label'],axis=1).sort_index(), test.drop(['label'],axis=1).sort_index())
```

```python
train, test = feature_engineering(sales,evals)
```

```python
import konlpy
import re 
from konlpy.tag import Twitter
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Komoran
from konlpy.tag import Kkma
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from soynlp.word import WordExtractor
from soynlp.utils import DoublespaceLineCorpus
from soynlp.tokenizer import LTokenizer
from soynlp.utils import DoublespaceLineCorpus
from soynlp.tokenizer import MaxScoreTokenizer
from soynlp.noun import LRNounExtractor_v2
```

```python
text_data = pd.concat([sales['상품명'],evals['상품명']], axis=0).reset_index()
text_data = text_data.drop(['index'],axis=1)
```

```python
text_data
```

```python
corpus_path = text_data['상품명']
sents = DoublespaceLineCorpus(corpus_path,iter_sent=True)

word_extractor = WordExtractor(
    min_frequency=10, # example
    min_cohesion_forward=0.05,
    min_right_branching_entropy=0.0
)

word_extractor.train(corpus_path)
words = word_extractor.extract()

cohesion_score = {word:score.cohesion_forward for word, score in words.items()}
tokenizer = MaxScoreTokenizer(scores=cohesion_score)
```

```python
token_data = []
def token_pil(text,i):
    
    for i in range(i):
        nouns = tokenizer.tokenize(text[i])
        token_data.append(nouns)
        
    return token_data
    
    
token_data = token_pil(text_data['상품명'],len(text_data['상품명']))
```

```python
token_data
```

```python
num_features = 10    # 워드 백터 특정값 수
min_word_count = 1   # 단어에 대한 최소 빈도 수
num_workers = 4       # 프로세스 개수
context = 10         # 컨텍스트 윈도우 크기
downsampling = 1e-3   # 다운 샘플링 비율

from gensim.models import word2vec

print("Training model...")

model = word2vec.Word2Vec(token_data,
                        workers=num_workers,
                        size=num_features,
                        min_count=min_word_count,
                        window=context,
                        sample=downsampling)

def get_features(words, model, num_features):
    # 출력 벡터 초기화
    feature_vector = np.zeros((num_features), dtype=np.float32)
    
    num_words = 0
    # 어휘 사전 준비
    index2word_set = set(model.wv.index2word)
    
    for w in words:
        if w in index2word_set:
            num_words = 1
            # 사전에 해당하는 단어에 대해 단어 벡터를 더함
            feature_vector = np.add(feature_vector, model[w])
            
    # 문장의 단어 수만큼 나누어 단어 벡터의 평균값을 문장 벡터로 함
    feature_vector = np.divide(feature_vector, num_words)
    return feature_vector

def get_dataset(reviews, model, num_features):
    dataset = list()
    
    for s in reviews:
        dataset.append(get_features(s, model, num_features))
        
    reviewFeatureVecs = np.stack(dataset)
    return reviewFeatureVecs
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
# 예측이 제대로 이루어지지 않았다. 한 로우 당 60%가 틀림. 
# 왜 예측이 제대로 진행되지 않았는가??

# 1) 방송 노출 횟수가 너무 적었다. 정보가 없는데 학습을 시킨 느낌
# - 예측이 어찌되었든 시계열의 형태로 진행이 된다면
# - 누적 방송 노출 횟수같은 것도 변수의 일종이 될 수 있다. 

# 2) 아이템의 특성이 전혀 반영되지 않았다
# - 예측이 크게 빗나간 '도루코 페이스5 면도날 4개입' 경우 면도날이라는 특성 상 많이 사가는 사람이 없지 않을까? 
# - 이런 제품들은 한 번 사면 오래쓰는데? 단발적으로 사는 제품이 아니다. 이런 제품에 대한 특성도 변수로 기록될 수 있어야 한다. 
# - 과연 word2vec이 제대로 진행이 된걸까? -> 된 것 같긴함. 

# 3) 변수의 부족 
# - 날씨, 시청률 등의 변수를 부착해보자 


# 정리 
# 1) 변수 목록 - 날씨, 시청률, 브랜드 파워, 제품의 특성을 반영해주는 변수, 프라임 시간대, 생활물가지수 
# 2) 이상치 제거 - 예측이 하등 도움이 되지 않는 이상치를 제거하여 최대한 정규 분포의 형태로 만든다. 
# 3) 신기한게 여성 제품이 더 많이 팔리는 느낌이다. 
# 4) 농수축은 높은 확률로 high sale을 기록한다.


## 관련 자료 정리 

## http://news.bizwatch.co.kr/article/consumer/2020/06/08/0021 
# 홈쇼핑에 관한 기사 : 코로나19로 인해 홈쇼핑의 매출은 늘었으나 영업 이익은 줄어들었다.
#                   why? 고가 가전제품, 명품의 편성은 어려워지고 저수익 상품 판매가 늘었기 때문. 
#                   그렇다면 그렇게 크리티컬하게 상품 판매가 줄어들지는 않았을 것 같다. 

## https://byline.network/2019/11/08-21/ 
# 홈쇼핑 빅데이터 예측에 관한 기사 : 프라임 시간대 등 논리적인 접근을 통해 수상을 할 수 있었음. 
#                               빅콘에서도 논리적인 접근이 필요하다. 이 기사와는 차별점이 있는! 
```
