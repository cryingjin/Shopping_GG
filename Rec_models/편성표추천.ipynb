### 3 Wide & Deep Model 결과 확인 - 상품군 추천(2020년 1월1일)

```python
res = pd.read_csv('./res/recommend1.csv', encoding = 'cp949', index_col = 0)
```

```python
res['방송일시'] = pd.to_datetime(res['방송일시'])
res['방송월'] = res['방송일시'].dt.month
res['방송일'] = res['방송일시'].dt.day
res['방송시간(시간)'] = res['방송일시'].dt.hour
```

```python
res
```

```python
pred_list = after_schedule[after_schedule['방송일시'] < pd.to_datetime('2020-01-02 06:00:00')]
```

```python
pred_list['방송월'] = pred_list['방송일시'].dt.month
pred_list['방송일'] = pred_list['방송일시'].dt.day
pred_list['방송시간(시간)'] = pred_list['방송일시'].dt.hour
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
for args in res[['추천상품군1', '추천상품군2', '추천상품군3', '방송월', '방송일', '방송시간(시간)']].values:
    rec1, rec2, rec3 = args[:3]
    mon, day, hour = args[3:]

    value = pred_list.loc[(pred_list['방송월'] == mon) & (pred_list['방송일'] == day) & (pred_list['방송시간(시간)'] == hour) & (pred_list['상품군'].isin([rec1, rec2, rec3])), '취급액'] * 1.05
    pred_list.loc[(pred_list['방송월'] == mon) & (pred_list['방송일'] == day) & (pred_list['방송시간(시간)'] == hour) & (pred_list['상품군'].isin([rec1, rec2, rec3])), '취급액'] = value.values
```

```python
final_recommend_0101 = pred_list.sort_values(['방송월', '방송일', '방송시간(시간)', '취급액'], ascending = [True, True, True, False]).drop_duplicates(['방송월', '방송일', '방송시간(시간)'])
final_recommend_0101
```

```python
final_recommend_0101.to_excel(os.path.join('..', '쇼핑광고등어_편성표추천(2020.01.01).xlsx'), index = False)
```
