```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize']=(15,10)
sns.set(rc={'figure.figsize':(20,10)})
plt.style.use('seaborn') # seaborn 스타일로 변환
plt.rc('font', family='AppleGothic')
plt.rc('axes', unicode_minus=False)
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
# 데이터 업데이트
df = pd.read_excel(os.path.join('..', '0.Data', '01_제공데이터','2020 빅콘테스트 데이터분석분야-챔피언리그_2019년 실적데이터_v1_200818.xlsx'), skiprows = 1)
```

```python
df.head()
```

```python
plt.figure(figsize = (20, 7))
plt.subplot(1,2,1)
plt.plot(df['판매단가'].index, np.sort(df['판매단가'].values))
plt.title('판매단가 그래프', fontsize = 15)
plt.subplot(1,2,2)
plt.plot(df['판매단가'].index, np.sort(np.log(df['판매단가'].values)))
plt.title('판매단가 로그 그래프', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (20, 7))
plt.subplot(1,2,1)
plt.hist(df.loc[df['판매단가'] > 0, '판매단가'], bins = 50)
plt.title('판매단가 그래프', fontsize = 15)
plt.subplot(1,2,2)
plt.hist(np.log(df.loc[df['판매단가'] > 0, '판매단가'] + 1), bins = 50)
plt.title('판매단가 로그 그래프', fontsize = 15)
plt.show()
```
