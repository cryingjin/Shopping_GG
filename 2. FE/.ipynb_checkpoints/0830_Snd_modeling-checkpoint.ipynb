# 02. 모델링

```python
xgb_pars = {'min_child_weight': 1, 'eta': 0.05, 'colsample_bytree': 0.9, 
            'max_depth': 10, 'subsample': 0.9, 'lambda': 1. , 'booster' : 'gbtree', 'silent': 1,
            'objective': 'reg:linear', 'gamma' : 0.1}
```

```python
import random
import xgboost as xgb
from sklearn.metrics import mean_squared_error
model = xgb.XGBRegressor()
```

```python
val_preds = []
test_preds = []
for i in tqdm(range(1,13)):
    # 한달에서도 랜덤 샘플 반 뽑기
    test_idx = random.sample(list(X[X['방송월'] == i].index), int(len(X[X['방송월'] == 1]) / 2))
    train_idx = set(X.index) - set(test_idx)
    
    train = X.loc[train_idx]
    test = X.loc[test_idx]
    y_test = y.loc[test_idx]
    
    x_train = train[train['방송월'] != i]
    y_train = y.loc[train[train['방송월'] != i].index]
    
    x_val = train[train['방송월'] == i]
    y_val = y.loc[train[train['방송월'] == i].index]
    
    model.fit(x_train, y_train)
    val_pred = model.predict(x_val)
    pred = model.predict(test)
    
    val_preds.append(MAPE(y_val, val_pred))
    test_preds.append(MAPE(y_test, pred))
    print(np.sqrt(mean_squared_error(y_val, val_pred)), np.sqrt(mean_squared_error(y_test, pred)))
```

```python
def xgb_model(X, y):
    val_preds = []
    test_preds = []
    fe_info = dict()
    for k in X.columns:
        fe_info[k] = 0

    for i in range(1,13):
        # 한달에서도 랜덤 샘플 반 뽑기
        test_idx = random.sample(list(X[X['방송월'] == i].index), int(len(X[X['방송월'] == 1]) / 2))
        train_idx = set(X.index) - set(test_idx)

        train = X.loc[train_idx]
        test = X.loc[test_idx]
        y_test = y.loc[test_idx]

        x_train = train[train['방송월'] != i]
        y_train = y.loc[train[train['방송월'] != i].index]

        x_val = train[train['방송월'] == i]
        y_val = y.loc[train[train['방송월'] == i].index]

        dtrain = xgb.DMatrix(x_train, label = y_train)
        dvalid = xgb.DMatrix(x_val, label = y_val)

        watchlist = [(dtrain, 'train'), (dvalid, 'valid')]

        model = xgb.train(xgb_pars, dtrain, 10000, watchlist, early_stopping_rounds=100, verbose_eval=10)
        for k, v in model.get_fscore().items():
            fe_info[k] += v

        dtest = xgb.DMatrix(test)
        val_pred = model.predict(dvalid)
        y_pred = model.predict(dtest)

        val_preds.append(MAPE(y_val, val_pred))
        test_preds.append(MAPE(y_test, y_pred))
        for m, arg in enumerate(zip(val_preds, test_preds), 1):
            print(f'{m}월\t', '[val]:', arg[0], '\t[test]', arg[1])
        
    return val_preds, test_preds, fe_info
```

```python
val_preds = []
test_preds = []
fe_info = dict()
for k in X.columns:
    fe_info[k] = 0

for i in range(1,13):
    # 한달에서도 랜덤 샘플 반 뽑기
    test_idx = random.sample(list(X[X['방송월'] == i].index), int(len(X[X['방송월'] == 1]) / 2))
    train_idx = set(X.index) - set(test_idx)
    
    train = X.loc[train_idx]
    test = X.loc[test_idx]
    y_test = y.loc[test_idx]
    
    x_train = train[train['방송월'] != i]
    y_train = y.loc[train[train['방송월'] != i].index]
    
    x_val = train[train['방송월'] == i]
    y_val = y.loc[train[train['방송월'] == i].index]
    
    dtrain = xgb.DMatrix(x_train, label = y_train)
    dvalid = xgb.DMatrix(x_val, label = y_val)
    
    watchlist = [(dtrain, 'train'), (dvalid, 'valid')]
    
    model = xgb.train(xgb_pars, dtrain, 10000, watchlist, early_stopping_rounds=100, verbose_eval=10)
    for k, v in model.get_fscore().items():
        fe_info[k] += v
    
    dtest = xgb.DMatrix(test)
    val_pred = model.predict(dvalid)
    y_pred = model.predict(dtest)
    
    val_preds.append(MAPE(y_val, val_pred))
    test_preds.append(MAPE(y_test, y_pred))
    print(test_preds)
```

```python
plt.figure(figsize = (10, 7))
plt.plot(val_preds)
plt.plot(test_preds)
plt.xlabel('Month', fontsize = 14)
plt.title('MAPE', fontsize = 15)
plt.legend(['valid', 'test'])
plt.show()
```
