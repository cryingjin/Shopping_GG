```python
def MAPE(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
lgb_params1 = {'learning_rate': 0.03, 'max_depth' : 24, 'objective': 'regression', 'metric': 'mape', 'is_training_metric': True,
              'lambda_l2': 0.1, 'min_child_samples' : 40, 'n_estimators' : 7500, 'num_leaves' : 16, 'subsample' : 0.6}
```

```python
def lgbm_model(X, y, params) : 

    mape = {'val_mape' : [], 'test_mape' : []}
    
    # 실제 예측 값 
    preds = {'val_preds' : [], 'test_preds' : []}    
    lgbm_feat_imp = [0] * len(X.columns)

    
    for i in range(1,13):
        train_idx = X.loc[X['방송월'] != i].index
        test_idx = X.loc[X['방송월'] == i].index
        
        x_train = X.loc[train_idx]
        y_train = y.loc[x_train.index]
        
        test = X.loc[test_idx]
        
        x_val = test.loc[((test['방송일'] > 0) & (test['방송일'] < 8)) | ((test['방송일'] > 15) & (test['방송일'] < 24))]
        y_val = y.loc[x_val.index]
        
        x_test = test.loc[((test['방송일'] > 7) & (test['방송일'] < 16)) | ((test['방송일'] > 23) & (test['방송일'] < 32))]
        y_test = y.loc[x_test.index] 
        
        
        train_T = lgb.Dataset(x_train.values, label=y_train.values) 
        val_T = lgb.Dataset(x_val.values, label=y_val.values)        

        
        model = lgb.train(params, train_T, 10000, valid_sets = val_T, verbose_eval=500, early_stopping_rounds=500)
        val_pred = model.predict(x_val.values)
        y_pred = model.predict(x_test.values)  
        
        
        preds['val_preds'].append(np.exp(val_pred))
        preds['test_preds'].append(np.exp(y_pred))
        
        
        mape['val_mape'].append(MAPE(np.exp(y_val), np.exp(val_pred)))
        mape['test_mape'].append(MAPE(np.exp(y_test), np.exp(y_pred)))
        lgbm_feat_imp += model.feature_importance()        
        
        for m, arg in enumerate(zip(mape['val_mape'], mape['test_mape']), 1):
            print(f'{m}월\t', '[val]:', arg[0], '\t[test]', arg[1]) 
     
    
    return mape, preds, lgbm_feat_imp
```

## xgboost 

### ver 1

```python
xgb_params1 = {'min_child_weight': 1, 'eta': 0.03, 'colsample_bytree': 0.9, 'n_estimators' : 10000,
               'max_depth': 20, 'subsample': 0.9, 'lambda': 1. , 'booster' : 'gbtree', 'silent': 1,
               'objective': 'reg:linear', 'eval_metric' : 'mae', 'gamma' : 0.01}
```

```python
def xgb_model(X, y, params):
  
    mape = {'val_mape' : [], 'test_mape' : []}
    preds = {'val_preds' : [], 'test_preds' : []}    
    feat_imp = dict()
    for k in X.columns:
        feat_imp[k] = 0

    for i in range(1,13):
        train_idx = X.loc[X['방송월'] != i].index
        test_idx = X.loc[X['방송월'] == i].index
        
        x_train = X.loc[train_idx]
        y_train = y.loc[x_train.index]
        
        test = X.loc[test_idx]
        x_val = test.loc[((test['방송일'] > 0) & (test['방송일'] < 8)) | ((test['방송일'] > 15) & (test['방송일'] < 24))]
        y_val = y.loc[x_val.index]
        x_test = test.loc[((test['방송일'] > 7) & (test['방송일'] < 16)) | ((test['방송일'] > 23) & (test['방송일'] < 32))]
        y_test = y.loc[x_test.index]

        dtrain = xgb.DMatrix(x_train, label = y_train)
        dvalid = xgb.DMatrix(x_val, label = y_val)

        watchlist = [(dtrain, 'train'), (dvalid, 'valid')]

        model = xgb.train(params, dtrain, 10000, watchlist, early_stopping_rounds=500, verbose_eval=500)
        for k, v in model.get_fscore().items():
            feat_imp[k] += v

        dtest = xgb.DMatrix(x_test)
        val_pred = model.predict(dvalid)
        y_pred = model.predict(dtest)
        
        
        preds['val_preds'].append(np.exp(val_pred))
        preds['test_preds'].append(np.exp(y_pred))

        mape['val_mape'].append(MAPE(np.exp(y_val), np.exp(val_pred)))
        mape['test_mape'].append(MAPE(np.exp(y_test), np.exp(y_pred)))
        for m, arg in enumerate(zip(mape['val_mape'], mape['test_mape']), 1):
            print(f'{m}월\t', '[val]', arg[0], '\t[test]', arg[1])
        
    return mape, preds, feat_imp
```

# Ensemble 

### First 
* lgbm_ver2 + xgb_ver1 

```python
same_mth = [1, 12]
model1_mth = [2, 7, 10]
model2_mth = [3, 4, 5, 6, 7, 8, 9, 11]
```

```python
def ensemble_mape(model1, model2, same_mth, model1_mth, model2_mth, k) : 
    mape = {'val_mape' : [], 'test_mape' : []}
    
    
    idx_val = [X.loc[(X['방송월'] == i) & (((X['방송일'] > 0) & (X['방송일'] < 8)) | ((X['방송일'] > 15) & (X['방송일'] < 24)))].index for i in range(1,13)]
    idx_test = [X.loc[(X['방송월'] == i) & (((X['방송일'] > 7) & (X['방송일'] < 16)) | ((X['방송일'] > 23) & (X['방송일'] < 32)))].index for i in range(1,13)]
 

    for i in range(12) : 
        y_val, y_test = y.loc[idx_val[i]] , y.loc[idx_test[i]]
        
        if i+1 in same_mth : 
            val_pred = model1['val_preds'][i] * 0.5 + model2['val_preds'][i] * 0.5 
            test_pred = model1['test_preds'][i] * 0.5 + model2['test_preds'][i] * 0.5
            
        elif i+1 in model1_mth : 
            val_pred = model1['val_preds'][i] * k + model2['val_preds'][i] * (1-k)
            test_pred = model1['test_preds'][i] * k + model2['test_preds'][i] * (1-k)
            
        else : 
            val_pred = model1['val_preds'][i] * (1-k) + model2['val_preds'][i] * k
            test_pred = model1['test_preds'][i] * (1-k) + model2['test_preds'][i] * k
        
        
        mape['val_mape'].append(MAPE(y_val, val_pred))
        mape['test_mape'].append(MAPE(y_test, test_pred))
   

    for m, arg in enumerate(zip(mape['val_mape'], mape['test_mape']), 1):
        print(f'{m}월\t', '[val]', arg[0], '\t[test]', arg[1])       

        
    return mape
```
