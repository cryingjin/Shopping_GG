# Categorical CV 

```python
category = ['상품군_가구', '상품군_가전', '상품군_건강기능', '상품군_농수축', '상품군_생활용품', '상품군_속옷', '상품군_의류', '상품군_이미용', '상품군_잡화', '상품군_주방', '상품군_침구']
```

```python
import random

import warnings
warnings.filterwarnings(action='ignore')
```

```python
def MAPE(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
import lightgbm as lgb
```

```python
lgb_params2 = {'learning_rate': 0.02, 'max_depth' : 24, 'objective': 'regression', 'metric': 'mape', 'is_training_metric': True,
              'lambda_l1': 0.7, 'min_child_samples' : 35, 'n_estimators' : 10000, 'num_leaves' : 30, 'bagging_fraction': 0.8}
```

```python
def lgbm_model3(X, y, params) : 

    mape = {'category' : [], 'val_mape' : [], 'test_mape' : []}
    preds = {'category' : [], 'val_preds' : [], 'test_preds' : []}    
    lgbm_feat_imp = [0] * len(X.columns)
    
        
        
    for c in category : 
        print(c)
        
        train_idx, test_idx = X.loc[X[c] == 0].index, X.loc[X[c] == 1].index
        
        x_train, y_train = X.loc[train_idx], y.loc[train_idx]        
        x_test , y_test  = X.loc[test_idx] , y.loc[test_idx]
            
        
        mape_m = {'month' : [], 'val_mape' : [], 'test_mape' : []}
        pred_m = {'month' : [], 'val_pred' : [], 'test_pred' : []}
            
            
        for i in range(1,13) : 
            
            month = i
            
            x_train_m = x_train.loc[x_train['방송월'] != i]
            y_train_m = y_train.loc[x_train_m.index]        

            x_val_m = x_test.loc[x_test['방송월'] == i & (((x_test['방송일'] > 0) & (x_test['방송일'] < 8)) | ((x_test['방송일'] > 15) & (x_test['방송일'] < 24)))]
            y_val_m = y_test.loc[x_val_m.index]

            x_test_m = x_test.loc[x_test['방송월'] == i & (((x_test['방송일'] > 7) & (x_test['방송일'] < 16)) | ((x_test['방송일'] > 23) & (x_test['방송일'] < 32)))]
            y_test_m = y_test.loc[x_test_m.index]   
            

            
            if len(x_val_m) == 0 :  
                continue
                
            train_T = lgb.Dataset(x_train_m.values, label=y_train_m.values) 
            val_T = lgb.Dataset(x_val_m.values, label=y_val_m.values)       
                
            model = lgb.train(params, train_T, 10000, valid_sets = val_T, verbose_eval=1000, early_stopping_rounds=500)
            val_pred = model.predict(x_val_m.values)
            y_pred = model.predict(x_test_m.values)             
                
            pred_m['val_pred'].append(np.exp(val_pred))
            pred_m['test_pred'].append(np.exp(y_pred)) 
                
            mape_m['val_mape'].append(MAPE(np.exp(y_val_m), np.exp(val_pred)))
            mape_m['test_mape'].append(MAPE(np.exp(y_test_m), np.exp(y_pred)))
            lgbm_feat_imp += model.feature_importance() / (12*11)
                
                
            mape_m['month'].append(month)
            pred_m['month'].append(month)
                

        
        
        for m, arg in zip(mape_m['month'], zip(mape_m['val_mape'], mape_m['test_mape'])):
            print(f'{m}월\t', '[val]', arg[0], '\t[test]', arg[1]) 
        
        
        
        mape['category'].append(c)
        mape['val_mape'].append(mape_m['val_mape'])
        mape['test_mape'].append(mape_m['test_mape'])
        
        
        preds['category'].append(c)
        preds['val_preds'].append(pred_m['val_pred'])
        preds['test_preds'].append(pred_m['test_pred'])        
                    
    
    return mape, preds, lgbm_feat_imp
```

```python
# 홀수달만 나오는건 뭔 일 ... 
lgbm_mape2, lgbm_preds2, lgbm_feat_imp2 = lgbm_model2(X, y2, lgb_params2)
```

```python
# 진석님 코드 
def lgbm_model4(X, y, params):
    test_preds = []
    

    for i in range(1,13):
        train_idx = X.loc[X['방송월'] != i].index
        test_idx = X.loc[X['방송월'] == i].index
        
        train = X.loc[train_idx]
        test = X.loc[test_idx]
        
        month = {}
        for category in ['상품군_가구', '상품군_가전', '상품군_건강기능', '상품군_농수축', '상품군_생활용품', '상품군_속옷', '상품군_의류', '상품군_이미용', '상품군_잡화', '상품군_주방', '상품군_침구']:
            x_test = test.loc[test[category] == 1]
            y_test = y.loc[x_test.index]
            
            if len(x_test) == 0:
                continue
            x_train = train.loc[train[category] == 1]
            y_train = y.loc[x_train.index]
        
        
            train_T = lgb.Dataset(x_train.values, label=y_train.values) 
            model = lgb.train(params, train_T, 5000, verbose_eval=1000)

            y_pred = model.predict(x_test.values)
            month[category] = MAPE(np.exp(y_test), np.exp(y_pred))
            
        test_preds.append(month)
        
        
    for m, arg in enumerate(test_preds, 1):
        for k, v in arg.items():
            print(f'{m}월\t', f'{k}:', v)
            
        print(f'{m}월\t', 'test:', np.mean(list(arg.values())))
        
    return test_preds
```

```python
lgbm_preds3 = lgbm_model4(X, y2, lgb_params2)
```
