```python
import re
def make_copus_out(df, p_column_name = "상품명.1"):
    result_list = []
    for i in range(len(df)):
        tmp = []
        
        if type(df["브랜드"].iloc[i]) != float:
            tmp.extend([df["브랜드"].iloc[i]])
        if type(df[p_column_name].iloc[i]) != float:
            tmp_p = re.sub('[-=.#)/?:($}+]','',df[p_column_name].iloc[i])
            tmp.extend(tmp_p.split())
        if type(df["단위"].iloc[i]) != float:
            try :
                tmp_p = re.sub('[-=.#)/?:($}+]','',df["단위"].iloc[i])
                tmp.extend(tmp_p.split())
            except :
                tmp.extend(str(df["단위"].iloc[i]))
        
        if len(tmp) != 0:
            result_list.append(tmp)
    return result_list
```

```python
make_copus_out(df_MJ)
```

```python
make_copus_out(df_JS,"상품명다시")
```

```python
make_copus_out(df_YJ)
```

```python
make_copus_out(df_JB_2)
```

```python
make_copus_out(df_JB_1)
```

```python
corpus_ours = []
corpus_ours.extend(make_copus_out(df_JS,"상품명다시"))
corpus_ours.extend(make_copus_out(df_MJ))
corpus_ours.extend(make_copus_out(df_YJ))
corpus_ours.extend(make_copus_out(df_JB_2))
corpus_ours.extend(make_copus_out(df_JB_1))
```
