```python
# [취급액] 50000원인 상품 일단 제외
df2 = df.loc[df['취급액'] != 50000]
```

```python
# 마더코드를 동일한 상품이라고 가정했을 때,
df2.groupby('마더코드').agg({
    '취급액' : [('총판매액', np.sum)],
    '판매단가' : [('최저판매단가', np.min), ('최고판매단가', np.max)],
    '상품코드' : [('상품코드개수', 'nunique')]
}).sort_values(('취급액', '총판매액'), ascending = False)
```

```python
# 그런데 단순히 [마더코드] 가 동일한 상품이 완전 같은 상품을 의미하지는 않음
df2.loc[df2['마더코드'] == 100148].drop_duplicates('상품코드')
```

```python
# 그런데 단순히 [마더코드] 가 동일한 상품이 완전 같은 상품을 의미하지는 않음
df2.loc[df2['마더코드'] == 100845].drop_duplicates('상품코드')
```

```python
# 상품코드
```

```python
# 상품코드로 상품을 구분했을 때, 이 때는 '무이자 LG 울트라HD TV 65UK6800HNC' 와 '일시불 LG 울트라HD TV 65UK6800HNC' 도 다른 상품으로 인식
df2.groupby('상품코드').agg({
    '취급액' : [('총판매액', np.sum)],
    '판매단가' : [('최저판매단가', np.min), ('최고판매단가', np.max)],
}).sort_values(('취급액', '총판매액'), ascending = False)
```

```python
df2['방송일'] = df2['방송일시'].dt.date
df2['방송시간(시)'] = df2['방송일시'].dt.hour
df2['방송시간(분)'] = df2['방송일시'].dt.minute
df2['방송요일'] = df2['방송일시'].dt.weekday
```

```python
# 방송일시를 보면 20분단위로 업데이트 되고 있는데, 노출(분)이 None으로 중복으로 나오는 경우가 있음 -> 왜그럴까?
df2.head(50)
```

```python
# 요일에 따른 
plt.figure(figsize = (10, 6))
sns.countplot(df2['방송요일'])
plt.show()
```

```python
# 요일별 취급액
# 주말에 취급액(판매액)이 조금 더 높은 경우가 있음
df2.groupby('방송요일')['취급액'].sum().plot.bar(rot = 0, figsize = (10, 6))
plt.xticks([0,1,2,3,4,5,6], ['월', '화', '수', '목', '금' ,'토', '일'])
plt.show()
```

```python
# 시간대별 방송개수?
plt.figure(figsize = (12, 8))
sns.countplot(df2['방송시간(시)'])
plt.show()
```

```python
# 시간대별 취급액(판매액)
df2.groupby('방송시간(시)')['취급액'].sum().plot.bar(rot = 0, figsize = (12, 8))
plt.show()
```

```python
# 그러면 주말 혹은 방송이 많은 시간대에 높은 판매단가의 상품을 방송하지 않을까?
```

```python
# 주말이 평일보다 방송되는 상품의 판매단가가 높다고 판단되긴 함
f, ax = plt.subplots(1, 2, figsize=(17,7))
df2.groupby('방송요일')['판매단가'].mean().plot.bar(rot = 0, ax = ax[0])
df2.groupby('방송요일')['판매단가'].median().plot.bar(rot = 0, ax = ax[1])
ax[0].set_title('판매단가 평균')
ax[1].set_title('판매단가 중간')
plt.show()
```

```python
# 시간대에서 압도적으로 차이남
# 20시 이후 시간대의 방송에서 압도적으로 높은 가격의 상품을 방송에서 다룸
f, ax = plt.subplots(1, 2, figsize=(17,7))
df2.groupby('방송시간(시)')['판매단가'].mean().plot.bar(rot = 0, ax = ax[0])
df2.groupby('방송시간(시)')['판매단가'].median().plot.bar(rot = 0, ax = ax[1])
ax[0].set_title('판매단가 평균')
ax[1].set_title('판매단가 중간')
plt.show()
```
